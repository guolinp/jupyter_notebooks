```python
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal


y = []
for T in range(4):
    y.extend([0]*16)
    y.extend([1]*16)
x = np.array(range(len(y)))

for i in range(20):
    index = random.randrange(0, len(x))
    y[index] = 1 - y[index]
    
############################################################
plt.figure(figsize=(16,4))
plt.plot(x, y, marker = 'o')
plt.show()
```

```python
res = np.fft.fft(y)
real = abs(res.real)
imag = (res.imag)

freq = np.fft.fftfreq(x.shape[-1])

plt.figure(figsize=(16,4))
#plt.plot(freq, real, marker = 'o')
plt.plot(freq, imag, marker = 'o')
plt.show()
```

```python
#b,a = signal.butter(3,0.05,'highpass')
b,a = signal.butter(3,0.1,'lowpass')
sf = signal.filtfilt(b,a,y)
plt.figure(figsize=(16,4))
plt.plot(x,sf, marker='o', color='g')
plt.show()
```

```python
y = [1 if p > 0.5 else 0 for p in sf]

plt.figure(figsize=(16,4))
plt.plot(x, y, marker = 'o')
plt.show()
```

```python
res = np.fft.fft(y)
real = abs(res.real)
imag = (res.imag)

freq = np.fft.fftfreq(x.shape[-1])

plt.figure(figsize=(16,4))
#plt.plot(freq, real, marker = 'o')
plt.plot(freq, imag, marker = 'o')
plt.show()
```
